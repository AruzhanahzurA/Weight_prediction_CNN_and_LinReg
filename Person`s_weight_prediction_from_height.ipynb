```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
gender_stats= pd.read_csv('/content/drive/MyDrive/datasets/gender-weight_pounds-height_inches.csv')
gender=pd.DataFrame(gender_stats)
```

```python
import seaborn as sns
import plotly.express as px
```

```python
px.scatter(data_frame=gender, x='Height', y='Weight', color='Gender', opacity=0.7) #, trendline='ols', size='depth'
```

```python
#2
gender['Height']=gender['Height']*2.54
gender['Weight']=gender['Weight']* 0.453592
gender
```

```python
gender['Gender']=gender['Gender'].replace({'Male':int(0), 'Female':int(1)})
gender
```

```python
from sklearn.preprocessing import StandardScaler

norm_scaler = StandardScaler()
gender['Height'] = norm_scaler.fit_transform(gender[['Height']])
gender
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X = gender.drop(['Weight'], axis=1)
y = gender['Weight']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
y_train
```

```python
#             _train
import sklearn.linear_model as lm

lin_reg_model = lm.LinearRegression()
lin_reg_model.fit(X_train, y_train)

#        

y_train_pred = lin_reg_model.predict(X_train)
y_test_pred  = lin_reg_model.predict(X_test)

y_train_pred
```

```python
from sklearn.metrics import mean_squared_error
train_mse = mean_squared_error(y_train, y_train_pred, squared=True)
print("Train MSE: ", train_mse)
test_mse = mean_squared_error(y_test, y_test_pred, squared=True)
print("Test MSE: ", test_mse)
```

```python

new_data = {'Gender': [1, 0],
            'Height': [169, 190]}

new_data_df = pd.DataFrame(new_data)
new_data_df['Height'] = norm_scaler.transform(new_data_df[['Height']])

predictions = lin_reg_model.predict(new_data_df)
print(predictions)
```

```python
print("Coefficients:", lin_reg_model.coef_)
print("Intercept:", lin_reg_model.intercept_)
```
